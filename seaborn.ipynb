```python
import seaborn as sns
```

# Continuous data Analysis




1. distplot

2. joinplot

3. pairplot




```python
df=sns.load_dataset("tips")
df.head()
```

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

**biveriate Analysis**

JoinPlot

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex', height=5)
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg',height=5)
```

**Multivariate Analysis**

Pair plot

```python
sns.pairplot(df, height= 2)
```

```python
sns.pairplot(df,hue='sex', height=2)
```

**Univariate Analysis**

Dis plot

```python
sns.displot(df['tip'])
```

```python
sns.displot(df['tip'],kde=True)
```

# Catagorical Analysis



1. boxplot

2. violinplot

3. countplot

4. bar plot





```python
#count plot

sns.countplot(x='sex',data=df)
```

```python

sns.countplot(y='sex',data=df)
```

```python
#barplot

sns.barplot(x='sex',y='total_bill',data=df)
```

```python
#boxplot

sns.boxplot(x='smoker',y='total_bill', data=df)
```

```python
sns.boxplot(x="day", y="total_bill", data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,orient='v')
```

```python
# categorize my data based on some other categories

sns.boxplot(x="total_bill", y="day", hue="smoker",data=df)
```

```python
#violinplot

sns.violinplot(x="total_bill", y="day", data=df,palette='rainbow')
```
